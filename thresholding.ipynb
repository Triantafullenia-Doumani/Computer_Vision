```python

import numpy as np
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

#create grayscale image
def greyscale(threshold,img_in):
    
    #open and store the input image to np array
    B=np.array(Image.open(img_in))
    
    #run the np array of input photo to give new grayscale value to each pixel
    for y in range(len(B)):
        for x in range(len(B[y])):
            if len(B.shape)==3:#if the shape of input image is equal to 3 that means that the image if colorful
                               
                #store the avergare of blue red and green
                #if R=G=B that means thah the pixel is grayscale
                colorCode=(int(B[y][x][0])+int(B[y][x][1])+int(B[y][x][2]))/3
                                                                    
            else:
                colorCode=B[y][x] #store the value of pixel
            if colorCode>=threshold: #if the value of pixel is >= threshold then make it equal to white(255)
                B[y][x]=255
            else:
                B[y][x]=0 #if the value of pixel is < threshold then make it equal to black(0)

    plt.imshow(B, cmap="gray")
    plt.show()
    return;
print("threshold : 20")
greyscale(20,"trikoupi6.png")
print("threshold : 40")
greyscale(40,"trikoupi6.png")
print("threshold : 80")
greyscale(80,"trikoupi6.png")
print("threshold : 100")
greyscale(100,"trikoupi6.png")
print("threshold : 120")
greyscale(120,"trikoupi6.png")
print("threshold : 170")
greyscale(170,"color.png")
print("threshold : 180")
greyscale(180,"color.png")
print("threshold : 200")
greyscale(200,"color.png")
print("threshold : 230")
greyscale(230,"color.png")
print("threshold : 240")
greyscale(240,"color.png")
```
